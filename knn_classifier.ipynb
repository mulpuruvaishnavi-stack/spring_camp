```python
import numpy as np
```

```python
data = [
    [150, 7.0, 1, 'Apple'],
    [120, 6.5, 0, 'Banana'],
    [180, 7.5, 2, 'Orange'],
    [155, 7.2, 1, 'Apple'],
    [110, 6.0, 0, 'Banana'],
    [190, 7.8, 2, 'Orange'],
    [145, 7.1, 1, 'Apple'],
    [115, 6.3, 0, 'Banana']
]

# Label Encoding- bcs computers don't understand words, we need to give numbers.
label_map = {
    'Apple': 0,
    'Banana': 1,
    'Orange': 2
}

# Splitting into X and y
X = []
y = []

for row in data:
    X.append(row[:3])
    y.append(label_map[row[3]])

# Convert to NumPy arrays
X = np.array(X)
y = np.array(y)

print("X(Features):")
print(X)

print("\ny(Labels):")
print(y)
```

```python
def euclidean_distance(point1, point2):
    # Converting to numpy arrays
    point1 = np.array(point1)
    point2 = np.array(point2)
    diff = point1 - point2
    squared = diff ** 2
    sum_squared = np.sum(squared)
    distance = np.sqrt(sum_squared)

    return distance

    # Example: distance between first two fruits
d = euclidean_distance(X[0], X[1])
print("Distance:", d)
```

```python
class KNN:
    def __init__(self, k=3):
        self.k = k   # number of neighbors

    def fit(self, X, y):
        self.X_train = X
        self.y_train = y

    def euclidean_distance(self, point1, point2):
        return np.sqrt(np.sum((point1 - point2) ** 2))

    def get_distance(self,item):
          return item[0]

    def predict_one(self, x):
        distances = []

        # Step 1: Compute distance from all training points
        for i in range(len(self.X_train)):
            dist = self.euclidean_distance(x, self.X_train[i])
            distances.append((dist, self.y_train[i]))

        # Step 2: Sort by distance
        distances.sort(key=self.get_distance)

        # Step 3: Pick top k neighbors
        k_nearest = distances[:self.k]

        # Step 4: Voting
        labels = [label for (_, label) in k_nearest]
        prediction = max(set(labels), key=labels.count)

        return prediction

    def predict(self, X_test):
        predictions = []

        for x in X_test:
            pred = self.predict_one(x)
            predictions.append(pred)

        return np.array(predictions)
```

```python
# Test data
test_data = np.array([
    [118, 6.2, 0],  # Expected: Banana
    [160, 7.3, 1],  # Expected: Apple
    [185, 7.7, 2]   # Expected: Orange
])

# Create KNN model
model = KNN(k=3)

# Train (fit) the model
model.fit(X, y)

# Predict
predictions = model.predict(test_data)

print("Predicted labels (encoded):", predictions)
```

```python
# Expected labels (encoded)
y_expected = np.array([1, 0, 2])  # Banana, Apple, Orange

# predictions (already got in the above cell)
print("Predicted:", predictions)
print("Expected :", y_expected)

# Accuracy calculation
accuracy = np.sum(predictions == y_expected) / len(y_expected)
print("Accuracy:", accuracy)

k_values = [1, 3, 5]

for k in k_values:
    model = KNN(k=k)
    model.fit(X, y)
    predicts = model.predict(test_data)

    print(f"\nk = {k}")
    print("Predictions:", predicts)

# Create a modified dataset
X_modified = X.copy()
y_modified = y.copy()

# Change one training point (Apple which looks like Banana)
X_modified[0] = [120, 6.2, 1]

model = KNN(k=3)
model.fit(X_modified, y_modified)

preds_modified = model.predict(test_data)

print("\nPredictions after data change:", preds_modified)
```
